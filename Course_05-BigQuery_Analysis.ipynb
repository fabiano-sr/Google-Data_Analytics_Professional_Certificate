```python
import os
from google.cloud import bigquery
```

```python
# Definir a variável de ambiente
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'cursobigquery.json'

# Inicializalr o client
client = bigquery.Client()
```

```python
# Confirmar acesso ao Projeto
print("Projeto:", client.project)
```

#### <span style="color: gold;">Criação do DataSet</span>

```python
# Definir o nome do Dataset
dataset_name = "movie_data"

# Definir o Dataset_ID
dataset_id = f"{client.project}.{dataset_name}"
# Criar o objeto DataSet
dataset = bigquery.Dataset(dataset_id)
# Indicar a Localização
dataset.location = "US"

# Criar o DataSet
dataset = client.create_dataset(dataset, exists_ok=True)
```

```python
# Definir ID da tabela
table_id = f"{dataset_id}.movies"
# Indicar o local do dados de origem
csv_path = r'course5_asstes\\Course_05-Movie_Data.csv'

# Configurar o job de carregamento
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.CSV,
    autodetect=True,
)

# Criar e popular a tabela
with open(csv_path, "rb") as file:
    job = client.load_table_from_file(file, table_id, job_config=job_config)
```

```python
# Acessar a tabela
movies = client.get_table(f"{dataset_id}.movies")
```

#### <span style="color: gold;">Execussão de Querys</span>

```python
client.query(f'''
    SELECT 
        *
    FROM 
        {movies}
    WHERE
        Genre = "Comedy"
''').to_dataframe()
```
